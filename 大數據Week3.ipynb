```python
import pandas as pd
import numpy as np
from scipy.stats import chi2_contingency

data1 = [[315, 152, 337], [346, 126, 571]]
data2 = [[0.1, 0.15, 0.2, 0.25, 0.3, 0.35, 0.4, 0.45, 0.5, 0.55], [1, 0.95, 0.95, 0.9, 0.85, 0.7, 0.65, 0.6, 0.55, 0.42]]

stat1, p1, dof1, expected1 = chi2_contingency(data1)
covariance1 = np.cov(data1)
correlation1 = np.corrcoef(data1)

stat2, p2, dof2, expected2 = chi2_contingency(data2)
covariance2 = np.cov(data2)
correlation2 = np.corrcoef(data2)

print("covariance1 = ", covariance1)
print("\ncorrelation1 = ", correlation1)
print("\nstat1 = ", stat1)
print("P_value1 = ", p1)
print("degree of freedom1 = ", dof1)
print("expected1 = ", expected1)

print("\n\ncovariance2 = ", covariance2)
print("\ncorrelation2 = ", correlation2)
print("\nstat2 = ", stat2)
print("P_value2 = ", p2)
print("degree of freedom2 = ", dof2)
print("expected2 = ", expected2)
```
